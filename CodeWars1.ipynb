```python
def GetFactor(number):

    if (number == 2 or number == 3):
        return number

    if (number % 2 == 0):
        return 2
    if (number % 3 == 0):
        return 3

    divisor = 6
    while (divisor * divisor - 2 * divisor + 1 <= number):
        if (number % (divisor - 1) == 0):
            return divisor-1
        if (number % (divisor + 1) == 0):
            return divisor +1
        divisor += 6
    return number

def primeFactors(n):
    factors = []
    oldFactor = 0
    while n>1:
        factor = GetFactor(n)
        if(int(factor) != oldFactor):
            factors.append([int(factor),1])
            oldFactor = factor
        else:
            factors[-1][1]+=1
        n /= factor
        
            
    #Printing: -Works well-
    returnString = ""
    for i in range(len(factors)):
        if(factors[i][1]==1):
            returnString +='({0:0d})'.format(factors[i][0])
        else:
            returnString +='({0:0d}**{1:0d})'.format(factors[i][0],factors[i][1])
    return returnString
    
```

```python
import numpy as np
def MatMult(Mat1,Mat2):
    result = [[0,0],[0,0]]
    for i in range(2):
        for j in range(2):
            for k in range(2):
                result[i][j] += Mat1[i][k] * Mat2[k][j]
    return result


    
def MatPower (M, n):
    if n == 1:
        return M
    else:
        P = MatPower (M, int(n/2))
        if n%2 == 0:
            return MatMult(P,P)
        else:
            return MatMult(MatMult(P, P), M)
        
def fib(n):
    if n == 0:
        return 0
    if n == -1:
        return 1
    if n == 1:
        return 1
    forward = [[1,1],[1,0]]
    backward = [[0,1],[1,-1]]
    if n<0:
        n = abs(n)
        return MatPower(backward,n-1)[1][1]
    else:
        return MatPower(forward,n-1)[0][0]
    
    
```

```python
def peel(snail_map):
    return [row[1:-1] for row in snail_map[1:-1]]

def read_boarder(snail_map):
    text = []
    N = len(snail_map) #snail_map is NxN
    for i in range(N-1):
        text.append(snail_map[0][i])
    for i in range(N):
        text.append(snail_map[i][-1])
    for i in range(N-2,-1,-1):
        text.append(snail_map[-1][i])
    for i in range(N-2,0,-1):
        text.append(snail_map[i][0])
    return text

def snail(snail_map):
    text = []
    while len(snail_map)>=2:
        text= text + list(read_boarder(snail_map))
        snail_map = peel(snail_map)
    if len(snail_map)==1 and snail_map[0]:
        text.append(snail_map[0][0])
        
    return text
```

```python
def justify(text, width):
    listOfWords = text.split(" ")
    index = 0
    Script = ""
    while index < (len(listOfWords)):
        nextWords,LastSentence = pickNextWords(listOfWords[index:],width)
        SpaceLocations = findSpaces(nextWords, width,LastSentence)
        Sentence = makeSentence(nextWords,SpaceLocations)
        Script += Sentence 
        index += len(nextWords) 
    return Script.strip("\n")
```

```python
def makeSentence(nextWords,SpaceLocations):
    if (len(nextWords) == 1):
        return nextWords[0] + "\n"
    sentence = nextWords[0]
    for i in range(len(SpaceLocations)):
        sentence += " "*SpaceLocations[i]+nextWords[i+1]
    
    return sentence + "\n"
```

```python
def pickNextWords(listOfWords,width):
    sentenceLength = len(listOfWords[0])
    wordsInSentence = 1
    myBool = True
    lastSentence = False
    while (myBool == True):
        if (wordsInSentence < len(listOfWords)):
            None
        else:
            lastSentence = True
            break
            
        sentenceLength += len(listOfWords[wordsInSentence]) +1
        if (sentenceLength <= width):
            wordsInSentence +=1
        else:
            myBool = False
            
    return listOfWords[:(wordsInSentence)],lastSentence
    
```

```python
def findSpaces(sentenceWords, width, lastSentence):
    totalSpaces = width - sum(map(len,sentenceWords))
    nb_SpaceLocations = len(sentenceWords)-1
    if nb_SpaceLocations == 0:
        return []
    minSpacing = int(totalSpaces/nb_SpaceLocations)
    nb_extraSpacing = totalSpaces%nb_SpaceLocations
    SpaceLocations = []
    for i in range(nb_SpaceLocations):
        if lastSentence:
            SpaceLocations.append(1)
        else:
            if nb_extraSpacing >0:
                SpaceLocations.append(minSpacing+1)
                nb_extraSpacing -=1
            else:
                SpaceLocations.append(minSpacing)
    return SpaceLocations
    
```

```python
from collections import namedtuple

Dimensions = namedtuple("Dimensions", "columns rows")

def compute_product(matrixA, matrixB):
    matrixA_dim = Dimensions(len(matrixA[0]), len(matrixA))
    matrixB_dim = Dimensions(len(matrixB[0]), len(matrixB))
    if len(matrixA) != len(matrixB[0]):
        return "Error, Matrices cannot be multiplied due to dimensionality"
    product = []
    for col in range(matrixA_dim.columns):
        product.append([0 for row in range(matrixB_dim.rows)])
    for i in range(matrixA_dim.rows):
        for j in range(matrixB_dim.columns):
            product[i][j] += matrixA[i][j] * matrixB[j][i]
    return product
```

```python
def tower_builder(n_floors):
    block = "*"
    air = " "
    myTower = []
    for i in range(n_floors):
        myTower.append(air*(n_floors-i-1)+ block*(2*i+1)+air*(n_floors-i-1))
    return myTower
```

```python
def move_zeros(array):
    index = 0
    noZeroLength = len(array)
    while(index<noZeroLength):
        if (array[index] is 0) or (array[index] == 0.0 and type(array[index])==float):
            array[index:-1] = array[index+1:]
            array[-1] = 0
            noZeroLength-=1
        else:
            index +=1
    return array
```

```python
import numpy as np
def find_next_square(sq):
    if (pow(int(np.sqrt(sq)),2)==sq):
        return pow(int(np.sqrt(sq)+1),2)
    else: 
        return -1
```

```python
def disemvowel(string):
    vowels = {'a','e','i','o','u','A','E','I','O','U'}
    myString = ''
    for i in (string):
        if(not i in vowels):
            myString += i
    return myString
```
